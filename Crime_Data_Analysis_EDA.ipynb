#### 11. What was the overall percentage decrease in crime between 2020 and 2023?

```python
# Filter the data for the years 2020 and 2023
crime_2020 = df1[df1['Year'] == 2020]
crime_2023 = df1[df1['Year'] == 2023]

# Count the number of crimes in each year
crime_count_2020 = crime_2020.shape[0]
crime_count_2023 = crime_2023.shape[0]

# Calculate the percentage decrease
percentage_decrease = ((crime_count_2020 - crime_count_2023) / crime_count_2020) * 100

# Step 5: Print the result
print(f"The overall percentage decrease in crime between 2020 and 2023 is {percentage_decrease:.2f}%")
```

#### 12. Which counties had the highest number of vehicle thefts?

```python
# Filter the data for vehicle theft incidents
vehicle_thefts = df1[df1['Crm_Cd_Desc'].str.contains('VEHICLE', case=False, na=False)]

# Group by county
vehicle_thefts_by_county = vehicle_thefts.groupby('AREA_NAME').size().reset_index(name='Vehicle_Theft_Count')

# Sort the result in descending order to get the counties with the highest number of vehicle thefts
vehicle_thefts_by_county_sorted = vehicle_thefts_by_county.sort_values(by='Vehicle_Theft_Count', ascending=False)

print(vehicle_thefts_by_county_sorted)
```

#### 13. Which areas had the highest number of vehicle thefts?

```python
# Filter the dataset to get only 'Vehicle Theft' crimes
vehicle_theft = df1[df1['Crm_Cd_Desc'] == 'VEHICLE - STOLEN']

# Group by area and count the number of vehicle thefts
vehicle_theft_by_area = vehicle_theft.groupby('AREA_NAME')['DR_NO'].count().reset_index()

# Sort areas by number of vehicle thefts in descending order
vehicle_theft_by_area_sorted = vehicle_theft_by_area.sort_values(by='DR_NO', ascending=False)

# Display the areas with the highest number of vehicle thefts
print(vehicle_theft_by_area_sorted.head())
```
